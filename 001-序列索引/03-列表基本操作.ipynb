## 列表

直接使用 [] 进行创建

```python
lst = ['hello', 'world', 98, 100.5]
print(lst)
```

 使用内置函数 list() 创建列表

```python
lst2 = list('hello world')
lst3 = list(range(1,10,2)) # 1 到 10 ，步长 2
print(lst2)
print(lst3)
```

各种运算

```python
print(lst + lst2 + lst3)
print(lst * 3)
print(len(lst))
print(max(lst3))
print(min(lst3))
```

```python
# 竟然可以删除，wc

lst4 = [10,20,30]
print(lst4)
```

```python
del lst4
print(lst4)
```
