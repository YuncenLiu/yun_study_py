### 元组与列表的区别

元组
1. 不可变序列
2. 无法实现添加、删除和修改元素等操作
3. 支持切片访问，不支持修改操作
4. 访问和处理速度快
5. 可以作为字典的健

列表：
1. 可变序列
2. append、insert、remove、pop 等方法对元素操作
3. 支持切片访问和修改表中元素
4. 访问和处理速度慢
5. 不能作为字典的健

```python
t = ('hellp', [10,20,30], 'python', 'world')

print(t)
```

```python
# 内置 tuple() 创建元组
t = tuple('helloworld')
print(t)
```

```python
t = tuple([10,20,30,40])
print(t)
```

```python
print('10 在元组中是否存在', (10 in t))
print('10 是否在元素中不存在', (10 not in t))
print('最大值', max(t))
print('最小值', min(t))
print('len', len(t))
print('t.index', t.index(10))
print('t.count', t.count(10))
```

```python
t = (10)
print('type(t)', type(t))
```

如果元组中只有一个元素， 逗号不能省略，否则就成 int 类型了

```python
t = (10,)
print('type(t)', type(t))
```

## 元组的遍历与访问

```python
t = ('python', 'hello', 'world')
print(t[0])
print(t[1])
print(t[2])
```

```python
for item in t:
    print(item)
```

```python
for i in range(len(t)):
    print(i, '--->' ,t[i])
```

```python
for index,item in enumerate(t):
    print(index, '--->', item)
```

```python
for index,item in enumerate(t, 11):
    print(index, '--->', item)
```

### 元组生成式

```python
t = (i for i in range(1,4))
print(t)
```

```python
t = tuple(t)
print(t)
```

使用 __next__ 进行遍历，获取下一个元素，特性是，不管使用了next 还是 for 只要遍历过了，再用 next 就会拿不到值

```python
t = (i for i in range(1,4))
print(type(t))
# 只能处理生成器对象
print(t.__next__())
print(t.__next__())

# 后续再转成元组就只能获取剩余的元素
t = tuple(t)
print(t)
```
