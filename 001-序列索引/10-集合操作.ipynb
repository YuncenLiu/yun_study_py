### 列表、元组、字典、集合区别

列表 list， 可变序列，不重复，有序，[]

元组 tuple， 不可变序列，可重复，有序，()

字典 dict， 可变序列，key不可重复，value可以重复，无序，{key:vale}

集合 set， 不可变序列，不可重复，无序，{}

```python
A = {10, 20, 30, 40, 50}
B = {30, 50, 88 ,76, 20}

print(A&B)
```

```python
print(A|B)
```

```python
print(A-B)
```

```python
# 补集操作
print(A^B)
```

## 集合相关操作

```python
s = {10, 20, 30}
s.add(100)
print(s)
```

```python
s.remove(20)
print(s)
```

```python
s.clear()
print(s)
```

```python
s = {10, 20, 30}
for item in s:
    print(item)
```

```python
for index,item in enumerate(s):
    print(index, ' --> ', item)
```

```python
s = {i for i in range(1,10)}
print(s)
```
