### 序列和索引

序列是一个存储多个值的连续空间，每个值都有整数编号



```python
# 正向递增
s = 'helloworld'
for i in range(0, len(s)):
    print(i, s[i], end='\t\t')
```

```python
# 倒叙
s = 'helloworld'
for i in range(-10, 0):
    print(i, s[i], end='\t\t')
```

如上可知，索引也可以从负数开始

```python
s = 'helloworld'
print(s[9],s[-1], end='\t\t')
print(s[8],s[-2], end='\t\t')
```

### 切片操作的语法结构

![](./images/Snipaste_2025-06-19_00-04-52.png)

```python
s = 'helloworld'

s1 = s[0:5:2]
print(s1)

# 省略开始位置
s2 = s[:5:2]
print(s2)

# 省略步长
s3 = s[:5:]
print(s3)

# 省略结束位置
s4 = s[0::1]
print(s4)

# 省略结束
s5 = s[5:]
s6 = s[5::]
print(s5)
print(s6)
```
