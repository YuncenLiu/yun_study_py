### 序列相加操作

```python
s = 'hello'
s2 = 'world'

print(s + s2)
```

### 序列相乘操作

```python
print('>' * 20)
```

### 判断元素是否存在于字符串

```python
s = 'helloworld'
print('e在 helloworld 存在吗?', ('e' in s))
print('v在 helloworld 存在吗?', ('v' in s))

print('e在 helloworld 不存在吗?', ('e' not in s))
print('v在 helloworld 不存在吗?', ('v' not in s))
```

### 内置函数等

```python
print('len()',len(s))
print('max()',max(s))
print('min()',min(s))
```
