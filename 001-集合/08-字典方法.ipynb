### 字典方法

d.keys() 获取所有key数据

d.values() 获取所有 value 数据

d.pop(key,default) 可以存在获取相应的value，同时删除这个 kv, 否则获取默认值

d.popitem() 随机从字典中取一个 kv，结果为元组类型，然后再删除

d.clear() 清空字典中的所有 kv

```python
d = {1001: '李梅', 1002: '王华', 1003: '张峰'}
print(d)
```

```python
# 新增一个元素
d[1004] = '张丽丽'
print(d)
```

```python
keys = d.keys()
print(keys)
print(list(keys))
print(tuple(keys))
```

```python
# 获取字典中所有 value
values = d.values()
print(values)
print(len(values))
print(tuple(values))
```

```python
lst = list(d.items())
print(lst)

d = dict(lst)
print(d)
```

```python
print(d.pop(1001))
print(d)
```

```python
print(d.pop(1001, '不存在'))
print(bool(d))
```

```python
d.clear()
print(d)
```

每个 python 对象都有 bool 类型

```python
print(bool(d))
```

```python
import random
d = {item: random.randint(1,100) for item in range(4)}
print(d)
```

```python
lst = [1001,1002,1003]
lst2 = ['藏猫猫','胖嘟嘟','小叮当']
d = {key:value for key,value in zip(lst,lst2)}
print(d)
```
