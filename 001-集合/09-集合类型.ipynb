### 集合

1. Python 中的集合和数学集合概念一致
2. Python 中的集合是一个无序不重复元素序列
3. 集合中只能存储不可变数据类型
4. 在 Python 中集合使用 {} 创建
5. 与列表、字典一样，都是 Python 中的可变数据类型

```python
s = {'胖嘟嘟', '藏猫猫', '小叮当'}
print(s)
```

集合中只能存储不可变数据类型

```python
s = {[10,20], [30,40]}
print(s)
```

```python
s = set()
print(bool(s))
# 只有空才是 false
print(type(s))
```

```python
s = set('helloworld')
print(s)

s2 = set([10,20,30])
print(s2)

s3 = set(range(1,10))
print(s3)
```

```python
print('max', max(s3))
print('min', min(s3))
print('len', len(s3))
```
