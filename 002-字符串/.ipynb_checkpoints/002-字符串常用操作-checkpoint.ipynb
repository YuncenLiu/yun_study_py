## 字符串常用操作

字符串是 Python 中的不可变数据类型

str.lower(), 将 str 字符串全部转成小写，结果为新字符串

str.upper(), 将 str 字符串全部转成大写字母，结果为新字符串

str.split(sep = None) 把 str 按指定分隔符进行分割，结果为列表类型

str.count(sub) 结果为sub这个字符串在str中出现的次数

str.find(sub) 查询 sub 这个字符串在 str 中是否存在，如果不存在结果为 -1，否则结果为子字符串在 str 首次出现的索引

str.index(sub) 功能与 find 相同，区别在于要查询的子串 sub 不存在时，程序报错

str.startswith(s) 查询 str 是否以 s 开头，结果为布尔值

str.endswith(s) 查询 str 是否以 s 结尾，结果为布尔值


```python
s1 = 'HelloWorld'
new_s2 = s1.lower()
print(s1, new_s2)
```
