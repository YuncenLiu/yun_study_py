```python
lst = [88, 99 ,98, 00, 87]

for index in range(len(lst)):
    if str(lst[index])!='0':
        lst[index] = '19' + str(lst[index])
    else:
        lst[index] = '200' + str(lst[index])

print(lst)
```

使用 enumerate 函数

```python
lst = [88, 99 ,98, 00, 87]

for index,value in enumerate(lst):
    if str(lst[index]) != '0':
        lst[index] = '19' + str(lst[index])
    else:
        lst[index] = '200' + str(lst[index])

print(lst)
```
