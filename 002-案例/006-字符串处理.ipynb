### 正则表达式

* `.` 匹配任意字符（不包括换行符）
* `\w` 匹配任意字母、数字、下划线
* `\W` 匹配任意非字母、数字、下划线
* `\d` 匹配任意数字
* `\S` 匹配任意非空白字符
* `\s` 匹配任意空白字符
* `?` 匹配前面的字符零次或一次
* `*` 匹配前面的字符零次或多次
* `+` 匹配前面的字符一次或多次
* `^` 排除字符， [^0-9] 排除 0-9 字符
* `$` 匹配字符串结尾
* `{n}` 匹配n次
* `{n,m}` 匹配n到m次
* `{n,}` 匹配n次以上
* `[abc]` 匹配a、b、c中的任意一个

```python
import re
```

re.match(pattern, string, flags=0) 用于从字符串开头位置进行匹配，如果起始位置匹配成功，结果为 Match，否则为 None

re.search(pattern, string, flags=0) 用于在整个字符串中搜索第一个匹配的值，如果匹配成功为 Match 对象，否则为 None

re.findall(pattern, string, flags=0) 用于整个字符串搜索所有符合正则表达的值，结果为一个列表

re.sub(pattern, repl, string, count, flags=0) 用于实现对字符串指定子串进行替换

re.split(pattern, string, maxsplit, flags=0) 字符串中的 split() 方法相同，都是分割字符串

```python
import re
pattern='\d\.\d+' # + 限定了匹配最前面的字符串，最前面的字符串要包含数字，所以为 None
s = 'I study Python every Day'
# re.I 忽略大小写
match = re.match(pattern=pattern, string=s, flags=re.I)
print(match)
```

```python
s1 = '3.99 Python'
match_s1 = re.search(pattern=pattern, string=s1, flags=re.I)
print(match_s1)

print('匹配值的开始位置', match_s1.start())
print('匹配值的结束位置', match_s1.end())
print('匹配值的区间位置', match_s1.span())
print('待匹配的字符串', match_s1.string)
print('匹配的字符串', match_s1.group())
```
