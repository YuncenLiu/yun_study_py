## Python 中常用异常类型

ZeroDivisionError：当初数为0时，引发异常

IndexError：当索引越界时，引发异常

KeyError：当字典中不存在对应的键时，引发异常

NameError：当变量不存在时，引发异常

SyntaxError: 当语法错误时，引发异常

ValueError: 当字符串不能转为整数时，引发异常

AttributeError: 当对象没有对应的属性时，引发异常
```
int i = 10
print(i.name)
```

TypeError: 当对象类型错误时，引发异常

IndentationError: 当缩进错误时，引发异常


````python
try:
    num1 = int(input('请输入除数：'))
    num2 = int(input('请输入被除数：'))
    result = num1 / num2
    print('结果：', result)
except ZeroDivisionError:
    print('除数不能为0')
except ValueError:
    print('不能将在字符串转整数')
except BaseException:
    print('未知异常')
````

try...except...else

假设没有抛异常，则会执行 else 中的代码

````python
try:
    num1 = int(input('请输入除数：'))
    num2 = int(input('请输入被除数：'))
    result = num1 / num2
except ZeroDivisionError:
    print('除数不能为0')
except ValueError:
    print('不能将在字符串转整数')
else:
    print('结果：', result)
````

try...except...else...finally

是否有异常都会执行 finally

````python
try:
    num1 = int(input('请输入除数：'))
    num2 = int(input('请输入被除数：'))
    result = num1 / num2

except ZeroDivisionError:
    print('除数不能为0')
except ValueError:
    print('不能将在字符串转整数')
except BaseException:
    print('未知异常')
else:
    print('结果：', result)
finally:
    print('程序结束')
````

### 抛出异常

````python
try:
    gender = input('请输入性别：')
    if gender != '男' and gender != '女':
        raise ValueError('请输入正确的性别')
except ValueError:
    print('输入错误')
except BaseException:
    print('未知异常')
else:
    print('性别：', gender)
finally:
    print('程序结束')
````
