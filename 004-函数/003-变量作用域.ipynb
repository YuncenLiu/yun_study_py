## 变量作用域

是指变量起作用的范围，根据范围作用的大小可分为 局部变量 和 全局变量

局部变量： 定义在函数内，仅在函数内生效，函数执行完成后，局部变量的生命值也就结束了

全局变量：定义在函数外部，使用 global 关键字修饰的变量，作用于整个程序，

```python
def calc(a, b):
    s = a + b
    return s


result = calc(10 ,20)
print(result)
```

```python
a = 100 # 全局变量
def calc(x, y):
    return a + x + y

result = calc(10, 20)
print(result)
```

方法内修改全局变量

全局变量和局部变量同时在函数内，优先使用局部变量

```python
a = 100 # 全局变量
def calc(x, y):
    a = 200
    return a + x + y

result = calc(10, 20)
print(result)
print('a =', a)
```

函数内使用全局变量

```python
a = 100 # 全局变量
def calc(x, y):
    global a
    a = 200
    return a + x + y

result = calc(10, 20)
print(result)
print('a =', a)
```
