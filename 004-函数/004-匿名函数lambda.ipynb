### lambda 匿名函数

是指没有名字的函数，这种函数只能使用一次，一般是在函数的函数体只有一代码且只有一个返回值时，可以使用匿名函数来简化

#### 语法结构：

result = lambda 参数列表:表达式

```python
def calc(a, b):
    return a + b

print(calc(10 , 20))
```

```python
s = lambda a,b:a + b
# s 表示匿名函数

print(type(s))
print(s(10, 20))
```

使用匿名函数取列表函数值

```python
lst = [10, 20, 30, 40, 50]
for i in range(len(lst)):
    print(lst[i])
```

```python
for i in range(len(lst)):
    result = lambda x:x[i]
    print(result(lst))
```

排序

```python
student_scores = [
    {'name': '张三', 'score': 75},
    {'name': '李四', 'score': 80},
    {'name': '王五', 'score': 90},
    {'name': '赵六', 'score': 85}
]

student_scores.sort(key=lambda x:x.get('score'), reverse=True)
print(student_scores)
```
