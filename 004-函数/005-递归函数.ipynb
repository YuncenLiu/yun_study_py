## 递归函数

在一个函数的函数体内调用函数本身，就是递归函数

```python
def fuc(n):
    if n==1:
        return 1
    else:
        return n * fuc(n - 1)

print(fuc(5))
print(5 * 4 * 3 * 2 * 1)
```

斐波那契数列（黄金分割线）又称为生兔子数列

f(n) = f(n-1) + f(n-2)

```python
def fuc(n):
    if n == 1 or n == 2:
        return 1
    else:
        return fuc(n-1) + fuc(n-2)

for i in range(1, 20):
    print(fuc(i), end=', ')
```
