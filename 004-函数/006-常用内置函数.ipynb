## 常用的内置函数

1. bool(obj) 获取指定对象的 obj 布尔值
2. str(obj) , 获取指定对象 obj 转成字符串类型
3. int(x) , 将 x 转成 int 类型
4. float(x) , 将 x 转成 float 类型
5. list(sequence) , 将序列转成列表类型
6. tuple(sequence) ,  将序列转成元组类型
7. set(sequence) , 将序列转成集合类型

```python
print('非空字符串布尔值', bool('hello'))
print('空字符串布尔值', bool(''))
print('非零数字布尔值', bool(1))
print('零数字布尔值', bool(0))
print('None 布尔值', bool(None))

print('-' * 30)

print('列表布尔值', bool([1, 2, 3]))
print('元组布尔值', bool((1, 2, 3)))
print('字典布尔值', bool({'name': '张三'}))
print('集合布尔值', bool(set([1, 2, 3])))

print('-' * 30)

print('空列表布尔值', bool([]))
print('空元组布尔值', bool(()))
print('空字典布尔值', bool({}))
print('空集合布尔值', bool(set()))
```

```python
print(int(20) + int('40'))

print(float(90) + float('40'))
```

## 常用的数学函数

1. abs(x) 返回 x 的绝对值
2. divmod(x, y) 返回 x/y 的商和余数
3. max(x1, x2, ...) 返回 x1, x2, ... 中的最大值
4. min(x1, x2, ...) 返回 x1, x2, ... 中的最小值
5. sum(iter) 返回 iter 的和
6. pow(x, y) 函数返回 x 的 y 次方
7. round(x, n) 返回 x 四舍五入到 n 位的数
